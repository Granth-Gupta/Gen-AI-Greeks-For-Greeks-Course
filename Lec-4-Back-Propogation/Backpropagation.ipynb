```python
train_images, test_images, train_lables, test_lables = train_test_split(
    mnist_images, mnist_lables, test_size=0.2, random_state=42)
```

```python
# Activation Function
def relu(ws):
  return np.maximum(0, ws)

def softmax(ws):
  exp_ws = np.exp(ws - np.max(ws, axis=-1, keepdims=True))
  return exp_ws / np.sum(exp_ws, axis=-1, keepdims=True)
```

```python
# Layer
def dense_layer_neurons(input, weights, bias, activation='relu'):
  weighted_sum = np.dot(input, weights) + bias
  return relu(weighted_sum) if activation == 'relu' else softmax(weighted_sum)
```

```python
# Hyperparameters

imput_size = 28*28  # number of pixels in the image
hidden_size = 128   # Can be anything
output_size = 10    # number of classes

batch_size = 64
learning_rate = 0.1
epochs = 10
steps = len(train_images) // batch_size
```

```python
# Initialize Weights
weights_input_hidden = np.random.randn(imput_size, hidden_size) * 0.01
bias_input_hidden = np.zeros((1, hidden_size))

weights_hidden_output = np.random.randn(hidden_size, output_size) * 0.01
bias_hidden_output = np.zeros((1, output_size))

weights_hidden_output = np.random.randn(hidden_size, output_size) * 0.01
bias_hidden_output = np.zeros((1, output_size))
```

```python
# The Training Loop

for epoch in range(epochs):
  for step in range(steps):
    start_idx = step * batch_size
    end_idx = (step + 1) * batch_size

    # Forward Pass
    input_data = train_images[start_idx:end_idx]
    hidden_layer = dense_layer_neurons(input_data, weights_input_hidden, bias_input_hidden, activation='relu')
    output_layer = dense_layer_neurons(hidden_layer, weights_hidden_output, bias_hidden_output, activation='softmax')

    # Compute Loss - categorical crossentropy
    loss = -np.sum(train_lables[start_idx:end_idx] * np.log(output_layer + 1e-8)) / batch_size

    # Backward Pass (Gradient Descent)
    delta = output_layer - train_lables[start_idx:end_idx]
    gred_hidden = np.dot(delta, weights_hidden_output.T) * (hidden_layer > 0)  # Differenciation using matrix multiplication

    # Update weights and baises
    weights_hidden_output = weights_hidden_output - learning_rate * np.dot(hidden_layer.T, delta) / batch_size
    bias_hidden_output = bias_hidden_output - learning_rate * np.sum(delta, axis=0, keepdims=True) / batch_size

    weights_input_hidden = weights_input_hidden - learning_rate * np.dot(input_data.T, gred_hidden) / batch_size
    bias_input_hidden = bias_input_hidden - learning_rate * np.sum(gred_hidden, axis=0, keepdims=True) / batch_size

  # Compute Accuracy on validation set
  hidden_layer_val = dense_layer_neurons(test_images, weights_input_hidden, bias_input_hidden)
  output_layer_val = dense_layer_neurons(hidden_layer_val, weights_hidden_output, bias_hidden_output, activation='softmax')
  predictions = np.argmax(output_layer_val, axis=1)
  true_labels = np.argmax(test_lables, axis=1)
  accuracy = np.mean(predictions == np.argmax(test_lables, axis=1))

  print(f'Epoch {epoch + 1}/{epochs}, Loss: {loss:.4f}, Accuracy: {accuracy:.4f}')
```
