## Operations on Arrays

### Reshaping Arrays

```python
arr = np.arange(8)
print(arr)
print(arr.shape) # (8,)
reshaped_arr = arr.reshape((2, 2, 2))
print(reshaped_arr)
print(reshaped_arr.shape)

# convert any array into 1D (also called flattening of array)
print(reshaped_arr.reshape(-1)) # very, very common thing a data scientist will do
```

### Element-wise Addition

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result = arr1 + arr2
print(result)
```

```python
# Scalar Addition
print(arr1 + 20) # np.array([1,2,3]) + np.array([20,20,20])
# Scalar Multiplication
print(arr1 * 20) # np.array([1,2,3]) * np.array([20,20,20])
# Scalar Subtraction
print(arr1 - 20)
# Scalar Division
print(arr1 / 20)
```

### Concatenation

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])
print(arr1.shape, arr2.shape)
concatenated = np.concatenate((arr1, arr2), axis=0) # concatenate a row
print(concatenated)
```

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 8], [6, 9]]) # notice the change in arr2 shape
print(arr1.shape, arr2.shape)
concatenated = np.concatenate((arr1, arr2), axis=1) # concatenate a column
print(concatenated)
```

### Boolean Indexing

```python
arr = np.array([1, 2, 3, 4, 5])
# Create a boolean array where the condition is true
condition = arr > 3
# np.array([1, 2, 3, 4, 5]) > np.array([3, 3, 3, 3, 3])
print(condition)
print(type(condition))


# Use this boolean array to filter the original array
filtered_arr = arr[condition]
print(filtered_arr)
```

```python
arr = np.array([1, 2, 3, 4, 5])
arr[arr > 3]
```

## Broadcasting

Broadcasting is a powerful mechanism that allows NumPy to perform operations on arrays of different shapes.

### Broadcasting Example

```python
arr = np.array([1, 2, 3]) # np.array([[1,2,3], [1,2,3]])
arr2d = np.array([[1, 2, 3], [4, 5, 6]])
result = arr + arr2d # [[1+1, 2+2, 3+3], [1+4, 2+5, 3+6]]
print(result)
```

In this example, the 1D array arr is "broadcast" across each row of the 2D array arr2d.
